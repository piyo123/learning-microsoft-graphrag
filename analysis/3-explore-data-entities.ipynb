```python
# like: SELECT type, count(*) as cnt FROM entities GROUP BY type ORDER BY type
# df.sort_values("type").groupby("type").size() # NG -> 結果は返るがgroupbyの結果をSortしているわけではない。
# pandasは順序を保持するが、groupbyなどした後は順序が保証されないので、明示的にソートする必要がある。
# groupbyは既定でソートする。
# df.groupby("type", sort=True).size().sort_index() と同意
# groupby した type列は Index なので、sort_valueではなく、sort_indexを使う必要がある。
df.groupby("type").size().sort_index()
```
